```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import my_functions_tp02 as mf
import tools

plt.rcParams['image.cmap'] = 'gray'
plt.rcParams['figure.figsize'] = [5.0, 5.0]
plt.rcParams['figure.dpi'] = 140
```

```python
im_gray = cv2.imread('./figs/flower.png', cv2.IMREAD_GRAYSCALE)
# im_gray = cv2.imread('/content/drive/MyDrive/CoursBTI/figs/flower.png', 0)
plt.imshow(im_gray)
```

Appliquon le stride sans padding (2,5,11) et (1,3,5) 

```python
strides= [(2,2),(5,5),(11,11)]
dilatations= [(1,1),(3,3),(5,5)]
# dim kernel = 3
dimK = 3
kernel = np.ones((dimK,dimK)) * 1/dimK
fig , axs = plt.subplots(1, 3, figsize=(9, 3))
axs[0].imshow(tools.Conv2D(im_gray,kernel=kernel,stride=strides[0],dilation=dilatations[0]))
axs[0].set_title(f'dimK = {dimK}\nstride = {strides[0][0]} dilatation = {dilatations[0][0]}')

axs[1].imshow(tools.Conv2D(im_gray,kernel=kernel,stride=strides[1],dilation=dilatations[1]))
axs[1].set_title(f'dimK = {dimK}\nstride = {strides[1][0]} dilatation = {dilatations[1][0]}')

axs[2].imshow(tools.Conv2D(im_gray,kernel=kernel,stride=strides[2],dilation=dilatations[2]))
axs[2].set_title(f'dimK = {dimK}\nstride = {strides[2][0]} dilatation = {dilatations[2][0]}')

# dim kernel = 5
dimK = 5
kernel = np.ones((dimK,dimK)) * 1/dimK
fig , axs = plt.subplots(1, 3, figsize=(9, 3))
axs[0].imshow(tools.Conv2D(im_gray,kernel=kernel,stride=strides[0],dilation=dilatations[0]))
axs[0].set_title(f'dimK = {dimK}\nstride = {strides[0][0]} dilatation = {dilatations[0][0]}')

axs[1].imshow(tools.Conv2D(im_gray,kernel=kernel,stride=strides[1],dilation=dilatations[1]))
axs[1].set_title(f'dimK = {dimK}\nstride = {strides[1][0]} dilatation = {dilatations[1][0]}')

axs[2].imshow(tools.Conv2D(im_gray,kernel=kernel,stride=strides[2],dilation=dilatations[2]))
axs[2].set_title(f'dimK = {dimK}\nstride = {strides[2][0]} dilatation = {dilatations[2][0]}')

# dim kernel = 11
dimK = 11
kernel = np.ones((dimK,dimK)) * 1/dimK
fig , axs = plt.subplots(1, 3, figsize=(9, 3))
axs[0].imshow(tools.Conv2D(im_gray,kernel=kernel,stride=strides[0],dilation=dilatations[0]))
axs[0].set_title(f'dimK = {dimK}\nstride = {strides[0][0]} dilatation = {dilatations[0][0]}')

axs[1].imshow(tools.Conv2D(im_gray,kernel=kernel,stride=strides[1],dilation=dilatations[1]))
axs[1].set_title(f'dimK = {dimK}\nstride = {strides[1][0]} dilatation = {dilatations[1][0]}')

axs[2].imshow(tools.Conv2D(im_gray,kernel=kernel,stride=strides[2],dilation=dilatations[2]))
axs[2].set_title(f'dimK = {dimK}\nstride = {strides[2][0]} dilatation = {dilatations[2][0]}')

```
